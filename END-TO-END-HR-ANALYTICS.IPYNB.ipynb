```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
```

```python
np.random.seed(0)
n = 1000

df = pd.DataFrame({
    "EmployeeID": range(1, n+1),
    "Age": np.random.randint(18, 60, n),
    "Gender": np.random.choice(["Male", "Female"], n),
    "Department": np.random.choice(["HR", "Sales", "Research & Development"], n),
    "JobRole": np.random.choice([
        "Manager", "Sales Executive", "Research Scientist",
        "Laboratory Technician", "Human Resources", "Developer"
    ], n),
    "MonthlyIncome": np.random.randint(20000, 120000, n),
    "OverTime": np.random.choice(["Yes", "No"], n),
    "Attrition": np.random.choice(["Yes", "No"], n, p=[0.3, 0.7])
})
```

```python
total_employees = df["EmployeeID"].count()
attrition_count = df[df["Attrition"] == "Yes"].shape[0]
attrition_rate = (attrition_count / total_employees) * 100
avg_salary = df["MonthlyIncome"].mean()

print("Total Employees:", total_employees)
print("Attrition Count:", attrition_count)
print("Attrition Rate:", round(attrition_rate, 2), "%")
print("Average Salary:", int(avg_salary))
```

```python
plt.figure()
df["Attrition"].value_counts().plot.pie(autopct='%1.1f%%')
plt.title("Attrition Distribution")
plt.ylabel("")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df, y="JobRole", hue="Attrition")
plt.title("Attrition by Job Role")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df, x="Department", hue="Attrition")
plt.title("Attrition by Department")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x="OverTime", hue="Attrition")
plt.title("Attrition vs Overtime")
plt.show()
```
